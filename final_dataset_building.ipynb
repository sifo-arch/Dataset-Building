```python
import glob
import os
import pandas as pd
from pathlib import Path
import re
```

```python
AI_FOLDER = "ai"
HUMAN_DATASET = 'human-written-text-dataset.csv'
```

```python
def get_models_folders(ai_folder):
    pattern = os.path.join(ai_folder, '*')
    models = glob.glob(pattern,)
    return models

models_folders = get_models_folders(AI_FOLDER)
models_folders
```

```python
def get_ai_articles(model_folder):
    folder = Path(model_folder)
    def num_key(p: Path):
        n = re.search(r"(\d+)", p.stem)
        return int(n.group(1))
    articles_files = sorted(folder.glob("*.txt"), key=num_key)
    # print(articles_files[:10])
    # return
    articles = []
    for art_file in articles_files:
        with open(art_file, 'r', encoding='utf8') as file:
            articles.append(file.read())
    return articles

gemini_articles = get_ai_articles(models_folders[0])
gpt_articles = get_ai_articles(models_folders[1])
mistral_articles = get_ai_articles(models_folders[2])
```

```python
gpt_labels = ['GPT' for _ in range(len(gpt_articles))]
gemini_labels = ['Gemini' for _ in range(len(gemini_articles))]
mistral_labels = ['Mistral' for _ in range(len(mistral_articles))]
```

```python
articles = []
articles.extend(gpt_articles)
articles.extend(gemini_articles)
articles.extend(mistral_articles)

labels = []
labels.extend(gpt_labels)
labels.extend(gemini_labels)
labels.extend(mistral_labels)
print(labels)

ai_articles = pd.Series(articles, name='ai')
labels = pd.Series(labels, name='generated by')
```

```python
df = pd.read_csv(HUMAN_DATASET)
df
```

```python
df = df.assign(ai=ai_articles)
df
```

```python
df = df.assign(generated_by=labels)
df
```

```python
df.to_csv("dataset.csv", index=False)
```
